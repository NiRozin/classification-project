# Seleção de características

Verificou-se que as características 'education' e 'education_num', são redundantes e como a segunda é numérica que apresenta um valor ordinal, optou-se por manter a segunda.

```python
df[["education", "education_num"]].groupby(["education", "education_num"]).count().reset_index().sort_values("education_num")
```

```python
df.drop(columns="education", inplace=True)
```

```python
df.describe()
```

Para estado civil e relacionamento existe uma relação, mas não é única, por exemplo, não aparece esposa e esposo para quando o estado civil é de não casado: divorciado, separado, etc. Optou-se por não descartar previamente nenhuma delas.

```python
df[["marital_status", "relationship"]].groupby(["marital_status", "relationship"]).count().reset_index()
```

As variáveis _capital_gain_ e _capital_loss_ representam o capital positivo e negativo, respectivamente, dessa forma optou-se por usar uma única variável com ambos os valores diferenciando apenas como sinal positivo/negativo.

```python
df[df.capital_gain > 0].describe()
```

```python
df[df.capital_loss > 0].describe()
```

```python
df["capital_variation"] = df.capital_gain - df.capital_loss
```

```python
df.drop(columns=["capital_gain", "capital_loss"], inplace=True)
```

Verificou-se pela documentação do conjunto de dados que a variável _fnlwgt_ c onsidera um peso para nacionalidade da pessoa entre outras coisas. 
Como a variável _native_country_ é bem desbalanceada, optou-se por utilizar apenas para identificar quem possue nacionalidade Americana.

```python
df[["native_country", "fnlwgt"]].groupby("native_country").std()
```

```python
df["is_american"] = np.where(df.native_country == ' United-States', 1, 0)
```

```python
df.drop(columns=["native_country"], inplace=True)
```
